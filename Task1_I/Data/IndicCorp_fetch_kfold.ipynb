```python
import os
import pandas as pd
from sklearn.model_selection import StratifiedKFold

BASE_PATH = '/content/drive/MyDrive/svnit_shared_task/shared_task/bhasha-workshop/Task1_I/Data/IndicCorp_fetch_aug_label'

languages = {
    "bangla": "ben_Beng",
    "hindi": "hin_Deva",
    "malayalam": "mal_Mlym",
    "tamil": "tam_Taml",
    "telugu": "tel_Telu"
}

N_FOLDS = 5

for lang_name, lang_code in languages.items():
    print(f"\nðŸ“˜ Language: {lang_name.upper()} ({lang_code})")
    print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")

    # Paths
    lang_folder = os.path.join(BASE_PATH, lang_name)
    csv_file = os.path.join(lang_folder, 'Data_fetch_aug_label.csv')
    kfold_folder = os.path.join(lang_folder, 'K_Fold')

    if not os.path.exists(csv_file):
        print(f"âŒ File not found: {csv_file}")
        continue

    os.makedirs(kfold_folder, exist_ok=True)

    # Load CSV
    df = pd.read_csv(csv_file)
    texts = df["Input Sentences"]
    labels = df["Grammatical Error"]

    skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=42)

    for fold_idx, (train_index, test_index) in enumerate(skf.split(texts, labels), start=1):
        fold_folder = os.path.join(kfold_folder, f'fold_{fold_idx}')
        os.makedirs(fold_folder, exist_ok=True)

        # Split
        X_train = texts.iloc[train_index]
        X_test = texts.iloc[test_index]
        y_train = labels.iloc[train_index]
        y_test = labels.iloc[test_index]

        # Save CSVs
        X_train_file = os.path.join(fold_folder, 'X_train.csv')
        X_test_file  = os.path.join(fold_folder, 'X_test.csv')
        y_train_file = os.path.join(fold_folder, 'y_train.csv')
        y_test_file  = os.path.join(fold_folder, 'y_test.csv')

        X_train.to_csv(X_train_file, index=False)
        X_test.to_csv(X_test_file, index=False)
        y_train.to_csv(y_train_file, index=False)
        y_test.to_csv(y_test_file, index=False)

        # Print Fold 1 details explicitly
        if fold_idx == 1:
            print(f"Fold 1:")
            print(f"  X_train: {len(X_train)} samples -> /{lang_name}/fold_1/X_train.csv")
            print(f"  X_test : {len(X_test)} samples -> /{lang_name}/fold_1/X_test.csv")
            print(f"  y_train: {len(y_train)} samples -> /{lang_name}/fold_1/y_train.csv")
            print(f"  y_test : {len(y_test)} samples -> /{lang_name}/fold_1/y_test.csv")

    print("All folds created successfully in:")
    print(f"/Task1_I/Data/IndicCorp_fetch_aug_label/K_Fold/{lang_name}/fold_1 to fold_5")
```
